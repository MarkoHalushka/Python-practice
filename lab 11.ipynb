```python
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = np.tan(x)

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_ylim(-10, 10)
ax.grid(True)

asymptotes = [-3*np.pi/2, -np.pi/2, np.pi/2, 3*np.pi/2]
for a in asymptotes:
    plt.axvline(x=a, color='red', linestyle='--')

tikzplotlib.save("tangent_plot.tex")

plt.show()
```
